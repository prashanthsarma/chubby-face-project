```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the Excel file into a DataFrame
df = pd.read_excel('data_v2.csv')
filtered_df = df[df['Chubby'] < 5]
```

```python
print(filtered_df.shape[0] * 2)
df.shape
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(filtered_df['Asc'],filtered_df['Moon'], filtered_df['Chubby'], marker='o', linestyle='-')
ax.title('Sales Over Time')
ax.xlabel('Date')
ax.ylabel('Sales')
ax.grid(True)
ax.show()
```
